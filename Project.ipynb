We plot a graph consisting of x axis as Average Land Temperatures and y axis to be the years. Before further analysis it is necessary to check if the time series data is stationary or not. Stationarity of the data is important in time series analysis because if the previous data follows some particular pattern there are high chances that it would follow the same in the future too. This feature makes forecasting simple and increases the accuracy. 

The stationarity of the data could be evaluated by plotting the graph and checking whether the mean and variance of average Land Temperature remain constant for all the years. Along with Stastical Properties of data we also consider Auto correlation, which must be zero.

Instead of observing the data graphically to determine the stationarity, we execute Dickey- Fuller's meathod. This method implies if it accepts null hypothesis then the data is non stationary. Otherwise, reject null hypothesis to conclude that the data is stationary. 

```python
from statsmodels.tsa.stattools import coint, adfuller
```

```python
def TestStationaryPlot(y):
    rol_mean = y.rolling(window = 12, center = False).mean()
    rol_std = y.rolling(window = 12, center = False).std()
    
    plt.plot(y, color = 'blue',label = 'Original Data')
    plt.plot(rol_mean, color = 'red', label = 'Rolling Mean')
    plt.plot(rol_std, color ='black', label = 'Rolling Std')
    plt.xticks(fontsize = 10)
    plt.yticks(fontsize = 10)
    
    plt.xlabel('Time in Years', fontsize = 25)
    plt.ylabel('Total Emissions', fontsize = 25)
    plt.legend(loc='best', fontsize = 25)
    plt.title('Rolling Mean & Standard Deviation', fontsize = 25)
    plt.show(block= True)
```

```python
def TestStationaryAdfuller(y, cutoff = 0.01):
    ts_test = adfuller(y, autolag = 'AIC')
    ts_test_output = pd.Series(ts_test[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    
    for key,value in ts_test[4].items():
        ts_test_output['Critical Value (%s)'%key] = value
    print(ts_test_output)
    
    if ts_test[1] <= cutoff:
        print("Strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root, hence it is stationary")
    else:
        print("Weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
TestStationaryPlot(y)
```

```python
TestStationaryAdfuller(y)
```

To Forecaste the Average Land Temperature Values, we use Seasonal ARIMA model which is an extension of ARIMA model.
ARIMA stands for Autoregressive Random Integrated Moving Average model. The model depends on three parameters, 'p' stands for number autoregressive terms. 
'q' stands for number of moving average terms.
'd' stands for number of defferences, there may be seasonal and trend differences which could be eliminated for better results in prediction.
SARIMAX model has extension to these parameters. It also consists of (P,Q,D). The purpose of each parameter is similar to (p,q,d) but (P,Q,D) calculate the terms for seasonal data. 

Using this model, we generate AIC factor which is necessary to measure how well the model would fit for the data for(p,d,q) and (P,D,Q) values. 
We prefer to consider the values of (p,d,q) and (P,D,Q) which have lowest AIC value, it implies the model created with this AIC value would use less number of features making the model less complex and ensure more accuracy. 

While using ARIMA model we generate p,d,q values using ACF and PACF graphs. The same can done through grid search in SARIMAX model.

```python

p = d = q = range(0, 2)

pdq = list(itertools.product(p, d, q))

seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
warnings.filterwarnings("ignore") 

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = mod.fit()

            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
model = sm.tsa.statespace.SARIMAX(y,
                                order=(1,1,1),
                                seasonal_order=(1,1,1,12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
result = model.fit()
print(result.summary().tables[1])
```

The next step is to plot model diagnostics in order to check whether the model has all the necessary properties to find best accuracy for the model. This plot is to check that theassumptions made by us in befor steps are perfectly correct. The histogram below tell that the mean is zero and variance is one, which means it follows normal distribution, that is important for forecasting. The q-q plot tells that the time series data is linear and follows a pattern. 

```python
result.plot_diagnostics(figsize=(16, 8))
plt.show()
```

VALIDATION 

In One Step ahead foreceast, we train the model with the entire data and predict for next year. The predicted values are verified with the already existing values in orderto find the accuracy. Have used Mean square errorin order to find the error in accuracy, that is 1-accuracy. 

Below we have trained the model upto 2012 and predicted for 2013. 
The graph represents that the model is working well with less managable error in accuracy. 

```python
pred = result.get_prediction(start=pd.to_datetime('2012-01-01'), dynamic=False)
pred_ci = pred.conf_int()
```

```python
ax = y['1970':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7,figsize=(20, 15))

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

ax.set_xlabel('Years')
ax.set_ylabel('Average Land Temperature')
plt.legend()

plt.show()
```

```python
y_forecasted = pred.predicted_mean
y_truth = y['2012-01-01':]

# Compute the mean square error
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
```
